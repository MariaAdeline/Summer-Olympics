```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
print("No of cities where summer olympics is held so far:", len(df['City'].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df.groupby('Sport')['Medal'].apply(lambda x:x[x=='Gold'].count()).sort_values(ascending=False).head().plot.bar(color='purple',figsize=(10,5))
```

### 3. Which sport is having most number of Medals so far? (Top 5)

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot.bar(color='orange',figsize=(10,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot.bar(color='blue',figsize=(10,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df.groupby('Athlete')['Medal'].apply(lambda x:x[x=='Gold'].count()).sort_values(ascending=False).head().plot.bar(color='red',figsize=(10,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
yr=df.loc[(df['Country']=='IND') & (df['Medal']=='Gold'),['Year']].sort_values(by='Year',ascending=True).values[0][0]
print('India won the first Gold medal in Summer Olympics in the year',yr)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df.groupby('Event').count()['Athlete'].sort_values(ascending=False).head().plot.bar(color='green',figsize=(10,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df=df[df['Gender']=='Women']
df=df[df['Medal']=='Gold']
df.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot.bar(color='brown',figsize=(10,5))
```
